```python
from pyspark.sql import SparkSession
from pyspark.sql.functions import col, mean, stddev, variance, min, max, expr
import matplotlib.pyplot as plt

# Initialize a Spark session
spark = SparkSession.builder \
    .appName("SummaryStatistics") \
    .getOrCreate()



#  summary statistics
summary = cleaned_df.select(
    min(col('X1')).alias('min_X1'),
    max(col('X1')).alias('max_X1'),
    mean(col('X1')).alias('mean_X1'),
    expr('percentile_approx(X1, 0.5)').alias('median_X1'),
    variance(col('X1')).alias('variance_X1'),
    stddev(col('X1')).alias('stddev_X1')
)

# Show summary statistics
summary.show()

# Collect 'X1' values
x1_values = cleaned_df.select('X1').rdd.flatMap(lambda x: x).collect()

# Generate a histogram
plt.figure(figsize=(10, 6))
plt.hist(x1_values, bins=20, color='skyblue', edgecolor='black')
plt.xlabel('X1 Values')
plt.ylabel('Frequency')
plt.title('Histogram of X1 Feature')
plt.grid(True)
plt.show()
```

```python
from pyspark.sql.functions import col

# the number of rows where 'X1' is greater than 50 and 'Y1' equals 1
count_filtered_rows = cleaned_df.filter((col('X1') > 50) & (col('Y1') == 1)).count()

print("Number of rows where 'X1' is greater than 50 and 'Y1' equals 1:", count_filtered_rows)
```

```python
from pyspark.ml.feature import VectorAssembler
from pyspark.ml.classification import LogisticRegression, RandomForestClassifier
from pyspark.ml.evaluation import BinaryClassificationEvaluator
from pyspark.ml import Pipeline
from pyspark.sql.functions import col

# Ensure 'Y1' column contains only two unique labels (0 and 1)
cleaned_df = cleaned_df.filter(col("Y1").isin([0, 1]))

# Split the data into training and testing sets
train_data, test_data = cleaned_df.randomSplit([0.8, 0.2], seed=42)

# Filter out non-numeric columns
numeric_cols = [col for col in cleaned_df.columns if col != 'Y1' and cleaned_df.select(col).dtypes[0][1] != 'string']

# Define the feature vector assembler
assembler = VectorAssembler(inputCols=numeric_cols, outputCol="features")

# Define classification models
lr = LogisticRegression(labelCol="Y1", featuresCol="features")
rf = RandomForestClassifier(labelCol="Y1", featuresCol="features")

# Build pipelines for the models
lr_pipeline = Pipeline(stages=[assembler, lr])
rf_pipeline = Pipeline(stages=[assembler, rf])

# Fit models to the training data
lr_model = lr_pipeline.fit(train_data)
rf_model = rf_pipeline.fit(train_data)

# Evaluate model performance on the test data
lr_predictions = lr_model.transform(test_data)
rf_predictions = rf_model.transform(test_data)

# Evaluate AUC for logistic regression
lr_evaluator = BinaryClassificationEvaluator(labelCol="Y1")
lr_auc = lr_evaluator.evaluate(lr_predictions)

# Evaluate AUC for random forest
rf_evaluator = BinaryClassificationEvaluator(labelCol="Y1")
rf_auc = rf_evaluator.evaluate(rf_predictions)

print("Logistic Regression AUC:", lr_auc)
print("Random Forest AUC:", rf_auc)
```

```python
from pyspark.sql import SparkSession
from pyspark.sql.functions import col
from pyspark.ml.feature import VectorAssembler
from pyspark.ml.classification import LogisticRegression, RandomForestClassifier
from pyspark.ml.evaluation import MulticlassClassificationEvaluator

# Initialize SparkSession
spark = SparkSession.builder \
    .appName("Classification_Model_Comparison") \
    .getOrCreate()

# Assuming 'df' is the DataFrame containing the dataset
# Convert columns to appropriate types
for col_name in df.columns:
    df = df.withColumn(col_name, df[col_name].cast("double"))

# Remove rows with null values
df = df.na.drop()

# Prepare the data: Split into features and target label
features_cols = df.columns[:-1]  # All columns except the last one (target label)
assembler = VectorAssembler(inputCols=features_cols, outputCol="features")
data = assembler.transform(df).select("features", "Y1")

# Split the data into training and testing sets (70% train, 30% test)
train_data, test_data = data.randomSplit([0.7, 0.3], seed=42)

# Train Logistic Regression model
log_reg = LogisticRegression(labelCol="Y1", featuresCol="features")
log_reg_model = log_reg.fit(train_data)

# Train Random Forest model
rf = RandomForestClassifier(labelCol="Y1", featuresCol="features")
rf_model = rf.fit(train_data)

# Evaluate the models on the testing data
evaluator = MulticlassClassificationEvaluator(labelCol="Y1", predictionCol="prediction", metricName="accuracy")

# Evaluate Logistic Regression model
log_reg_pred = log_reg_model.transform(test_data)
log_reg_accuracy = evaluator.evaluate(log_reg_pred)

# Evaluate Random Forest model
rf_pred = rf_model.transform(test_data)
rf_accuracy = evaluator.evaluate(rf_pred)

# Print the performance of each model
print("Logistic Regression Model Accuracy:", log_reg_accuracy)
print("Random Forest Model Accuracy:", rf_accuracy)

# Make conclusions
if log_reg_accuracy > rf_accuracy:
    print("Logistic Regression model performs better.")
elif log_reg_accuracy < rf_accuracy:
    print("Random Forest model performs better.")
else:
    print("Both models have the same accuracy.")

# Stop Spark session
spark.stop()
```

```python
#  DataFrame as a temporary view
df_without_X10.createOrReplaceTempView("data")

# SQL query to select 'X2' and 'X8' columns where conditions are met
result = spark.sql("""
    SELECT X2, X8
    FROM data
    WHERE X2 > 1.0 AND X8 > 70
""")

# result
result.show()
```

```python
from pyspark.ml.regression import LinearRegression
from pyspark.ml.feature import VectorAssembler
from pyspark.ml.evaluation import RegressionEvaluator

# features vector
assembler = VectorAssembler(inputCols=["X2"], outputCol="features")
data = assembler.transform(df_without_X10).select("features", "X8")

# Split the data into training and testing sets
train_data, test_data = data.randomSplit([0.8, 0.2], seed=42)

# Train linear regression model
lr = LinearRegression(featuresCol="features", labelCol="X8")
lr_model = lr.fit(train_data)

# Make predictions on the test data
predictions = lr_model.transform(test_data)

# Evaluate the model
evaluator = RegressionEvaluator(labelCol="X8", predictionCol="prediction", metricName="rmse")
rmse = evaluator.evaluate(predictions)

# Print the RMSE (Root Mean Squared Error)
print("Root Mean Squared Error (RMSE):", rmse)
```

```python
from pyspark.ml.regression import LinearRegression
from pyspark.ml.feature import VectorAssembler
from pyspark.ml.evaluation import RegressionEvaluator

# Assemble features vector
assembler = VectorAssembler(inputCols=[col for col in df_without_X10.columns if col != 'X8'], outputCol="features")
data = assembler.transform(df_without_X10).select("features", "X8")

# Split the data into training and testing sets
train_data, test_data = data.randomSplit([0.8, 0.2], seed=42)

# Train Lasso regression model
lasso = LinearRegression(featuresCol="features", labelCol="X8", elasticNetParam=1.0, regParam=0.1)
lasso_model = lasso.fit(train_data)

# Make predictions on the test data
predictions = lasso_model.transform(test_data)

# Evaluate the model
evaluator = RegressionEvaluator(labelCol="X8", predictionCol="prediction", metricName="rmse")
rmse = evaluator.evaluate(predictions)

# Print the RMSE (Root Mean Squared Error)
print("Root Mean Squared Error (RMSE):", rmse)
```
